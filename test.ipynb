```python
donations_donor_proj = donations[['_projectid', '_donor_acctid', '_donationid']]
d = donations_donor_proj.sort_values([ '_projectid', '_donor_acctid'])
d
```

```python
count = 0
project_donornum = {}
dd = d.shift(1)

for index, row in d.iterrows():
    if (row['_donor_acctid'] != dd.iloc[index]['_donor_acctid']):
        key = row['_projectid']
        if key in project_donornum:
            project_donornum[key] += 1
        else:
            project_donornum[key] = 1
    else:
        print('same donor on different project!')
project_donornum
```

```python
donor_map = donations_donor_proj.groupby(['_projectid', '_donor_acctid']).count()
donor_map
```

```python
donor_count = donor_map.groupby('_donor_acctid').count() # single donor donate different project number 
donor_count['_donationid'].max()
```

```python
project_count = donor_map.groupby('_projectid').count() # project donated by different donors.
project_count['_donationid'].max()
```

```python
donor_count.groupby('_donationid')
```

```python
histo = donor_count['_donationid'].value_counts() 
histo = histo.sort_index()
# histo = histo[1:]
sum = histo.sum()
print(histo)
print(histo.iloc[0]/sum)
print(histo.iloc[1]/sum)
print(histo.iloc[2]/sum)
print(histo.iloc[3]/sum)
print(histo.iloc[4]/sum)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
# histo.hist(bin=10)
plt.hist(histo, bins=int(len(histo.index)/20))
plt.show()
```
